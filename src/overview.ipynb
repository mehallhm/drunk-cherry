```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from trail_helpers import load_trail_data, clean_trails, KNOWN_DIFFICULTIES
```

```python
# read in the data
trails = load_trail_data("../all_trails.zip")
trails = clean_trails(trails)
trails.head()
```

```python
np.arange(len(trails["difficulty"].unique()))
```

```python
# histogram of trail difficulty
plt.figure(figsize=(10, 4))
counts = trails["difficulty"].value_counts().reindex(KNOWN_DIFFICULTIES)
plt.bar(counts.index, counts)
plt.xlabel("Difficulty")
plt.ylabel("Number of Trails")
plt.title("Difficulty Counts")
plt.show()
```

```python
# Average number of GPX points
avg_length = trails['latitude'].apply(len).quantile(0.80)
print("Average number of GPX points", avg_length)
```
